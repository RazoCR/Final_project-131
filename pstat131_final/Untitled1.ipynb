```python
def plot_corr(df_corr,title,title2):
    f , ax=plt.subplots(figsize=(14,10))
    sns.heatmap(df_corr,ax=ax,cmap='coolwarm',linewidths=0,square=True,annot=True)
    plt.title(title)
    plt.xlabel(title2)
```

```python
plot_corr(dt.corr(),".",".")
```
